```python
import rospy
import threading
from std_msgs.msg import String

from eris.eris import Eris

from threading import Thread,Lock

from time import sleep
dataMutex=Lock()

#What to read from Eris
#Set up Eris driver to read these features


features=['saludo','Analog','Sine']
types=['char','float','float']
length=[4,5,5]
port='/dev/ttyACM0'

e=Eris(features,types,length,port)

pub = rospy.Publisher('chatt', String, queue_size=10)
#Data buffer
nums=[];

def publishData(data):
    #Make a local copy of the data
    dataMutex.acquire(1)
    local=list(data)
    dataMutex.release()
    a=e.parse(local)
    for i,sample in enumerate(a):
        print('Sample'+str(i))
        print(sample)


rospy.init_node('talker', anonymous=True)
rate = rospy.Rate(2) # 10hz 
e.sendCommand('S_ON') #Initilize automatic streaming of data
print('Inicio')
for i in range(0,10):
    #Get data from teensy
    dataMutex.acquire(1)    
    d=e.read()
    dataMutex.release()
    if len(d)>0:
        t=Thread(target=publishData,args=(d,))
        t.start()             
    rate.sleep()
e.sendCommand('S_OFF')
e.stop()

```
